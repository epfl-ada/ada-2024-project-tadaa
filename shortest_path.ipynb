```python
import pandas as pd
```

## Articles

```python
# Load the data
articles = pd.read_csv('wikispeedia_paths-and-graph/articles.tsv', sep = "\t", comment='#', header=None)
finished_paths = pd.read_csv("clean_finished_paths.csv")
unfinished_paths = pd.read_csv("clean_unfinished_paths.csv")
shortest_path = pd.read_csv('wikispeedia_paths-and-graph/shortest-path-distance-matrix.txt', sep = "\t", header=None) 
```

```python
articles.shape  == shortest_path.shape
print(articles.shape)
```

## Shortest Path distance matrix

```python
# Read line by line and store in a list
lines = []
with open('wikispeedia_paths-and-graph/shortest-path-distance-matrix.txt') as f:
    for line in f:
        if not (line.startswith('#') or line.startswith('\n')):
            # Convert space-separated values in the line to integers
            line_values = [int(c) if c != "_" else -1 for c in line.strip()]
            lines.append(line_values)

# Create a dictionary to map article names to indices
articles_dict = {}
for index, row in articles.iterrows():
    articles_dict[row[0]] = index
```

```python
# Calculate shortest path for each entry in finished_paths
finished_paths["shortest_path"] = finished_paths.apply(
    lambda x: lines[articles_dict[x["source"]]][articles_dict[x["target"]]],
    axis=1
)

# Calculate shortest path for each entry in unfinished_paths
unfinished_paths["shortest_path"] = unfinished_paths.apply(
    lambda x: lines[articles_dict[x["source"]]][articles_dict[x["target"]]],
    axis=1
)
```

```python
# Save the result
finished_paths.to_csv("clean_finished_paths.csv", index=False)
unfinished_paths.to_csv("clean_unfinished_paths.csv", index = False)
```
