```python
import pandas as pd
```

```python
# Read the data
articles = pd.read_csv('../wikispeedia_paths-and-graph/wikispeedia_paths-and-graph/articles.tsv', sep = "\t", header = None)
```

```python
#read line by line and store in a list
lines = []
with open('../wikispeedia_paths-and-graph/wikispeedia_paths-and-graph/shortest-path-distance-matrix.txt') as f:
    for line in f:
        lines.append(line)
```

```python
lines2 = []
for line in lines:
    l = []
    for c in line:
        if c == "_":
            c = -1
        if c == "\n":
            continue
        l.append(int(c))
    lines2.append(l[:-1])
print(lines2[0])
```

```python
articles
```

```python
# transform the data into a dictionary
articles_dict = {}
for index, row in articles.iterrows():
    articles_dict[row[0]] = index
print(articles_dict)
```

```python
finished_paths = pd.read_csv("./clean_finished_paths.csv")
finished_paths
```

```python
unfinished_paths = pd.read_csv("./clean_unfinished_paths.csv")
unfinished_paths
```

```python
finished_paths["shortest_path"] = finished_paths.apply(lambda x: lines2[articles_dict[x["source"]]-1][articles_dict[x["target"]]-1], axis = 1)
```

```python
unfinished_paths["shortest_path"] = unfinished_paths.apply(lambda x: lines2[articles_dict[x["source"]]-1][articles_dict[x["target"]]-1], axis = 1)
```

```python
finished_paths.to_csv("clean_finished_paths.csv", index = False)
unfinished_paths.to_csv("clean_unfinished_paths.csv", index = False)
```
