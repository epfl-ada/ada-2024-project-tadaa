```python
import pandas as pd
articles = pd.read_csv("wikispeedia_paths-and-graph/articles.tsv", sep="\t", header=None, skiprows=11)
articles
```

```python
from bs4 import BeautifulSoup

import os

def get_article_html_path(article_name):
    return os.path.abspath("./wikispeedia_articles_html/wpcd/wp/{}/{}.htm".format(article_name[0].lower(), article_name))

def get_first_paragraph(article_name):
    with (open(get_article_html_path(article_name), "r", encoding = "latin-1") as file):
        s = BeautifulSoup(file, "html.parser")
        first_paragraph = s.find("div", id="bodyContent")
        if first_paragraph is not None:
            first_paragraph = first_paragraph.find("p")
        else:
            first_paragraph = s.find("p")
        return first_paragraph.text.split("\n", 1)[0]

```

```python
dataset =  {}
from urllib.parse import unquote
from tqdm import tqdm
for article in tqdm(articles[0]):
    print(article)
    dataset[unquote(article)] = get_first_paragraph(article)
dataset
```

```python
import json
for explanation in dataset:
    dataset[explanation] = dataset[explanation].split(".")[0]
with open("explanations2.json", "w") as file:
    json.dump(dataset, file)
```
