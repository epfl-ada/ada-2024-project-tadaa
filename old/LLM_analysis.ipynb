```python
import json

from setuptools.command.rotate import rotate
from six import assertRaisesRegex
```

```python
# Load the data
with open('llm_responses.json') as f:
    responses = json.load(f)
    
responses = responses["source_target"]
```

```python
response_dict = {}
for response in responses:
    if response[0]+"_"+response[-1] not in response_dict:
        response_dict[response[0]+"_"+response[-1]] = []
    response_dict[response[0]+"_"+response[-1]].append(response)
response_dict
```

```python
import pandas as pd
finished_paths = pd.read_csv("clean_finished_paths.csv")
```

```python
grouped_by_source_target = finished_paths.groupby(["source", "target"]).count()
grouped_by_source_target
```

```python
sources = ["Asteroid", "Brain", "Theatre", "Pyramid", "Batman", "God"]
targets = ["Viking", "Telephone", "Zebra", "Bean", "The_Holocaust", "Terrorism"]
source_target_pairs = []
llm_means = []
people_means = []
for source, target in zip(sources, targets):
    source_target_pairs.append(source+"_"+target)
    selected_paths = finished_paths[(finished_paths["source"] == source) & (finished_paths["target"] == target)]
    mean = selected_paths["path_length"].median()
    people_means.append(mean)
    llm_mean = 0
    for l in response_dict[source+"_"+target]:
        llm_mean += len(l)
    llm_mean /= len(response_dict[source+"_"+target])
    llm_means.append(llm_mean)

print(llm_means)  
```

```python
import numpy as np
import matplotlib.pyplot as plt

# Define the width of the bars
bar_width = 0.35

# Set the positions of the bars on the x-axis
r1 = np.arange(len(source_target_pairs))
r2 = [x + bar_width for x in r1]

# Create the bars
plt.bar(r1, llm_means, color='b', width=bar_width, edgecolor='grey', label='LLM Means')
plt.bar(r2, people_means, color='r', width=bar_width, edgecolor='grey', label='People Mean')

# Add labels
plt.xlabel('Source -> Target', fontweight='bold')
plt.xticks([r + bar_width / 2 for r in range(len(source_target_pairs))], source_target_pairs, rotation=-45)

# Add legend
plt.legend()

# Show the plot
plt.show()
```

```python

asteroid_viking.sort_values("path_length")
# take th shortest paths
shortest_paths = asteroid_viking[asteroid_viking["path_length"] == asteroid_viking["path_length"].min()]
shortest_paths
```

```python
def remove_back_clicks(path):
    new_path = []
    count = 0
    for a in path[::-1]:
        if a == "<":
            count += 1
            continue
        if count == 0:
            new_path.append(a)
        else:
            count -= 1
    return new_path[::-1]
import ast
shortest_paths["path"] = shortest_paths["path"].apply(ast.literal_eval)
shortest_paths["cleaned_path"] = shortest_paths["path"].apply(remove_back_clicks)
shortest_paths["cleaned_path"]
```

```python
set_of_shortest_paths = set(shortest_paths["cleaned_path"].apply(str))
set_of_shortest_paths
```

```python
unfinished_paths["hour"] = unfinished_paths["timestamp"].apply(lambda x: x.split(" ")[1].split(":")[0])
finished_paths["hour"] = finished_paths["timestamp"].apply(lambda x: x.split(" ")[1].split(":")[0])
```

```python
#compare finished and unfinished paths by hour
finished_paths["hour"] = finished_paths["hour"].astype(int) - 5
unfinished_paths["hour"] = unfinished_paths["hour"].astype(int) - 5
if finished_paths["hour"].min() <= 0:
    finished_paths["hour"] = finished_paths["hour"] + 24
if unfinished_paths["hour"].min() <= 0:
    unfinished_paths["hour"] = unfinished_paths["hour"] + 24
unfinished_paths["hour"] %= 24
finished_paths["hour"] %= 24
finished_by_hour = finished_paths.groupby("hour").count()
unfinished_by_hour = unfinished_paths.groupby("hour").count()
# plot
plt.plot(finished_by_hour.index, finished_by_hour["path"], label="Finished Paths")
plt.plot(unfinished_by_hour.index, unfinished_by_hour["path"], label="Unfinished Paths")
plt.xlabel("Hour")
plt.ylabel("Number of Paths")
plt.title("Number of Paths by Hour")
plt.legend()
plt.show()
```
